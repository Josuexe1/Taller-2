```python
from scipy.optimize import newton
def f(x):
    y = x**3 - 3 * x**2 + x - 1
    print(f"f({x}) = {y}")
    return y

def f_prime(x):
    return 3 * x**2 - 6 * x + 1
estimations = []
def callback(x):
    estimations.append(x)
def newton_with_estimations(func, x0, fprime, tol=1e-6, maxiter=100):
    x = x0
    for _ in range(maxiter):
        callback(x)  # Guardar el valor actual
        fx = func(x)
        fpx = fprime(x)
        
        if abs(fpx) < 1e-10:  # Evitar divisin por cero
            raise RuntimeError("La derivada es cero")
        
        x_new = x - fx / fpx
        if abs(x_new - x) < tol:
            callback(x_new)  
            return x_new
        x = x_new
    
    raise RuntimeError("No se encontr la raz despus de las iteraciones mximas")

# Estimacin inicial
x0 = 100
root = newton_with_estimations(f, x0, f_prime)

print("Raz encontrada:", root)
print("Estimaciones visitadas:", estimations)
```

```python
from scipy.optimize import newton
import matplotlib.pyplot as plt

def f(x):
    return x**3 - 3 * x**2 + x - 1

def f_prime(x):
    return 3 * x**2 - 6 * x + 1

def newton_with_estimations(f, x0, f_prime, tol=1e-6, max_iter=100):
    points = []  
    x = x0
    for _ in range(max_iter):
        fx = f(x)
        fpx = f_prime(x)
        points.append((x, fx))  
        if abs(fx) < tol:
            break
        x -= fx / fpx  
    return x, points

# Valor inicial
x0 = 100
root, points = newton_with_estimations(f, x0, f_prime)

x_points = [point[0] for point in points]
plt.figure(figsize=(8, 6))
plt.plot(x_points, "o-", label="x points")
plt.grid(color="gray", linestyle="--", linewidth=0.5)
plt.title("Plot of x points in points")
plt.xlabel("Index")
plt.ylabel("x value")
plt.legend()
plt.show()
```
